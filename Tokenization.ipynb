```python
import nltk
```

```python
corpus = "Underneath the golden hue of the setting sun, the city bustled with life, each street a tapestry of stories waiting to unfold. Among the sea of faces, each harbored dreams, fears, and untold desires. In the quiet corners, secrets whispered like the rustle of autumn leaves, their murmurs lost in the cacophony of urban symphony. Shadows danced upon the walls, weaving tales of forgotten yesterdays and uncertain tomorrows. Amidst the chaos, a lone figure stood, searching for meaning in the labyrinth of existence. The air was heavy with anticipation, pregnant with the promise of change, of new beginnings. In the distance, sirens wailed, a haunting melody that underscored the fragility of life. Yet, amidst the turmoil, there lingered a sense of hope, a flickering flame that refused to be extinguished. And as night descended, the stars above bore witness to the eternal dance of humanity, a dance of joy, sorrow, and infinite possibility."
```

```python
corpus
```

```python
# Convert the paragraph into the sentences
from nltk.tokenize import sent_tokenize
```

```python
documents = sent_tokenize(corpus)
```

```python
# Convert the paragaraph to words
from nltk.tokenize import word_tokenize
```

```python
words = word_tokenize(corpus)
```

```python
words
```

```python
for sentences in documents:
    print(word_tokenize(sentences))
```

```python
from nltk.tokenize import TreebankWordTokenizer
```

```python
tokenizer= TreebankWordTokenizer()
```

```python
tokenizer.tokenize(corpus)
```
